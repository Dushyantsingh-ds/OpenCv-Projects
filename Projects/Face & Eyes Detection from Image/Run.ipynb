```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
#import libraries
```

```python

face_cascade = cv2.CascadeClassifier("Assets/Libraries/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("Assets/Libraries/haarcascade_eye.xml")
img = cv2.imread("Assets/Examples/dev.png")

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
#getting OpenCV lib & the image
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.show()
#apply so changes(grayscale)
```

```python
for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
#detection and drawing the shapes
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('my picture')
plt.show()
#show the output
```
